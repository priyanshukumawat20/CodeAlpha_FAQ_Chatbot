```python
faqs = {
    "What is CodeAlpha?": "CodeAlpha is a software development company that offers internship programs in fields like AI, Web Development, Data Science, and more.",
    "How long is the internship?": "The internship duration is 1 month.",
    "How many tasks do I need to complete?": "You need to complete a minimum of 2 to 3 tasks out of the 4 given tasks.",
    "Is the internship paid?": "No, this is an unpaid internship focused on learning and skill development.",
    "Will I get a certificate?": "Yes, you will receive a Completion Certificate which is QR verified, along with a Unique ID.",
    "How do I submit my tasks?": "You need to submit your completed tasks through the submission form shared in your WhatsApp group.",
    "Where should I upload my project code?": "You should upload your code to GitHub in a repository named CodeAlpha_ProjectName.",
    "Do I need to post on LinkedIn?": "Yes, you need to share your internship status and project video on LinkedIn, tagging @CodeAlpha.",
    "What happens if I submit only one task?": "Submitting only one task is considered incomplete, and certificates will not be issued.",
    "Will I get a Letter of Recommendation?": "Yes, a Letter of Recommendation is provided based on your performance.",
    "Is there any job opportunity after the internship?": "Yes, CodeAlpha offers job opportunities and placement support based on performance.",
    "How can I contact CodeAlpha?": "You can contact CodeAlpha via email at services@codealpha.tech or WhatsApp at +91 9336576683.",
    "Can I do the internship remotely?": "Yes, this is a remote internship and can be completed from anywhere.",
    "What tools or skills are needed for the AI internship?": "Basic knowledge of Python and willingness to learn AI/ML concepts is helpful for the AI internship.",
    "Is resume building support provided?": "Yes, CodeAlpha provides resume building support to interns."
}

print(f"Total FAQs loaded: {len(faqs)}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

questions = list(faqs.keys())
answers = list(faqs.values())

vectorizer = TfidfVectorizer()
question_vectors = vectorizer.fit_transform(questions)

def get_answer(user_question):
    user_vector = vectorizer.transform([user_question])
    similarities = cosine_similarity(user_vector, question_vectors)
    best_match_index = similarities.argmax()
    best_score = similarities[0][best_match_index]

    if best_score < 0.2:
        return "Sorry, I couldn't find a relevant answer. Please contact CodeAlpha at services@codealpha.tech"

    return answers[best_match_index]

# Test kar
print(get_answer("How long is the internship program?"))
```
